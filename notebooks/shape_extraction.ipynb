```python
import os
import matplotlib.pyplot as plt

from src.utils.loaders import load_images
from src.utils.shape_extraction import get_shape_mask_otsu, get_shape_mask_kmeans
```

```python
dataset_dir = './data/train/'
files = [os.path.join(dataset_dir, file) for file in os.listdir(dataset_dir)]
sample_size = 5

for path, image in load_images(files, sample_size):
    plt.imshow(image)
    plt.show()
```

```python
for path, image in load_images(files, sample_size):
    shape_mask_otsu = get_shape_mask_otsu(image)
    shape_mask_kmeans = get_shape_mask_kmeans(image)

    plt.figure(figsize=(10, 6))

    plt.subplot(1, 3, 1)
    plt.imshow(image)
    plt.title('Original image')

    plt.subplot(1, 3, 2)
    plt.imshow(shape_mask_otsu, cmap='gray')
    plt.title('Shape mask (Otsu)')

    plt.subplot(1, 3, 3)
    plt.imshow(shape_mask_kmeans, cmap='gray')
    plt.title('Shape mask (K-means)')

    plt.show()
```
